NÍVEL 1. "Calculadora" 

Exemplo: P(A|B) - Probabilidade de ter uma mutação (A) dado que o teste deu positivo (B)

```python
p_intersecao = 0.05 # 5% da população tem a mutação E teste positivo
p_condicao = 0.10 # 10% da população testou positivo no total

p_final = p_intersecao / p_condicao
print(f"A probabilidade condicional é: {p_final:.2f} ou {p_final*100}%")
```

NÍVEL 2. "Cientista de Dados" (NumPy & Pandas)

Na Bioinformática, raramente calculamos um número único. Trabalhamos com grandes tabelas (DataFrames). O Pandas é perfeito para aplicar a lógica daquela sua segunda imagem em milhares de linhas.

```python
import pandas as pd

# Simulando dados de pacientes
data = {
    'tem mutação': [True, True, False, True, False, True, True, False, True, True],
    'testou positivo': [True, False, True, True, False, True, False, False, True, False]
}
df = pd.DataFrame(data)

# Calculando P(tem mutação | testou positivo) automaticamente
positivo = df[df['testou positivo'] == True]
prob = positivo['tem mutação'].mean()

print(f"Probabilidade via Pandas: {prob:.2%}")
```

NÍVEL 3. "Bioinformática Avançada" (SciPy & Machine Learning)

Aqui é onde a probabilidade se torna poderosa. Você usará bibliotecas para:

    - Testes de Hipótese (P-value) --> Usar o scipy.stats para saber se a diferença de expressão de um gene entre um grupo doente e saudável é fruto do acaso ou real.

    - Classificadores Naive Bayes --> Um algoritmo de Machine Learning (disponível no scikit-learn) que se baseia inteiramente em Probabilidade Condicional para classificar sequências de DNA ou prever riscos de doenças.

```python
import numpy as np
from sklearn.naive_bayes import GaussianNB

# 1. Preparando os dados (Exemplo: 0 = Saudável, 1 = Portador, 2 = Condição)
# As 'features' podem ser a presença de certas bases nitrogenadas ou SNPs
X = np.array([[0], [1], [2], [0], [2], [1], [2], [0]]) # Genótipos observados
y = np.array([0, 0, 1, 0, 1, 0, 1, 0])             # Condição clínica real

# 2. Criando o modelo baseado em Probabilidade Condicional (Naive Bayes)
modelo = GaussianNB()

# 3. Treinando o modelo
modelo.fit(X, y)

# 4. Predição: Qual a probabilidade de um novo paciente com genótipo '2' ser doente?
novo_paciente = [[2]]
probabilidades = modelo.predict_proba(novo_paciente)

print(f"Probabilidade de ser saudável: {probabilidades[0][0]:.2%}")
print(f"Probabilidade de ter a condição: {probabilidades[0][1]:.2%}")
```
