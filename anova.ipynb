```python
import pandas as pd
from scipy.stats import f_oneway

# Ler arquivo CSV
dados = pd.read_csv("dados_anova.csv")

# Separar os Grupos
controle = dados[dados["grupo"] == "Controle"]["expressao"]
mutante  = dados[dados["grupo"] == "Mutante"]["expressao"]
heterozigoto = dados[dados["grupo"] == "Heterozigoto"]["expressao"]

# Rodando ANOVA
F, p = f_oneway(controle, mutante, heterozigoto)

# Mostrando os Resultados
print("--- RESULTADO ANOVA ---")
print("-----------------------")
print(f"F-statistic: {F}")
print(f"p-value: {p}")

# Interpretação
if p < 0.05:
    print("\nExiste diferença significativa entre os grupos.")
else:
    print("\nNão existe diferença significativa entre os grupos.")
```
