DESAFIO 01 - Teste T de duas amostras

```python
import numpy as np
from scipy import stats

# 1. Definição das Hipóteses:
# H0 (Nula): Não há diferença entre os grupos (médias iguais).
# H1 (Alternativa): A terapia gênica aumenta a hemoglobina fetal.

# 2. Dados Simulados (Níveis de HbF após 6 meses)
controle = [1.2, 1.5, 1.1, 1.8, 1.3, 1.6, 1.4, 1.2]
terapia_genica = [4.5, 5.2, 3.8, 6.1, 4.9, 5.5, 4.2, 5.8]

# 3. Execução do Teste-t de Student (Amostras Independentes)
t_stat, p_valor = stats.ttest_ind(controle, terapia_genica)

# 4. Resultados
print("--- Resultado da Análise Estatística ---")
print(f"Estatística T: {t_stat:.4f}")
print(f"p-valor: {p_valor:.8f}")

# 5. Tomada de Decisão (Alpha = 0.05)
if p_valor < 0.05:
    print("Decisão: Rejeitar a Hipótese Nula (H0).")
    print("Conclusão: A terapia gênica teve um efeito significativo no aumento da HbF.")
else:
    print("Decisão: Falha em rejeitar a Hipótese Nula (H0).")
    print("Conclusão: Não há evidências suficientes de que a terapia funcione.")
```

DESAFIO 02 - Teste de Rigor (Alpha)

```python
import numpy as np
from scipy import stats

# Dados simulados: Níveis de hemoglobina fetal em 10 pacientes
# Grupo A: Placebo | Grupo B: Nova Terapia Gênica
controle = [1.2, 1.4, 1.3, 1.5, 1.2, 1.4, 1.3, 1.2, 1.1, 1.4]
terapia = [1.8, 1.9, 1.7, 2.0, 1.8, 1.9, 1.8, 1.7, 1.9, 2.0]

t_stat, p_valor = stats.ttest_ind(controle, terapia)

# Desafio: Altere o valor de 'limiar_rigoroso' para 0.00001 e veja o resultado
limiar_padrao = 0.05
limiar_rigoroso = 0.001 

print(f"p-valor calculado: {p_valor:.6f}")

if p_valor < limiar_rigoroso:
    print("Resultado: Significativo mesmo sob alto rigor!")
else:
    print("Resultado: Não é significativo para este nível de rigor.")
```

DESAFIO 03 - Teste Paredao (Análise Antes e Depois)

```python
import numpy as np
from scipy import stats

# Níveis de hemoglobina de 6 pacientes ANTES e DEPOIS do tratamento
antes = [1.2, 1.3, 1.1, 1.4, 1.2, 1.3]
depois = [2.5, 2.8, 2.4, 3.1, 2.7, 2.9]

# stats.ttest_rel é usado para amostras RELACIONADAS (mesmo indivíduo)
t_stat, p_valor = stats.ttest_rel(antes, depois)

print("--- Análise de Evolução do Paciente ---")
print(f"p-valor: {p_valor:.10f}")

if p_valor < 0.05:
    print("Conclusão: O tratamento causou uma mudança significativa no paciente.")
else:
    print("Conclusão: Não houve mudança estatisticamente relevante.")
```

DESAFIO 03 - Poder da Amostra (N pequeno)

```python
import numpy as np
from scipy import stats

# Cenário A: Apenas 2 pacientes (baixa confiança)
controle_pequeno = [1.2, 1.3]
terapia_pequeno = [2.5, 2.6]

# Cenário B: 10 pacientes com a mesma média (alta confiança)
controle_grande = [1.2, 1.3] * 5
terapia_grande = [2.5, 2.6] * 5

_, p_pequeno = stats.ttest_ind(controle_pequeno, terapia_pequeno)
_, p_grande = stats.ttest_ind(controle_grande, terapia_grande)

print(f"p-valor com amostra pequena (n=2): {p_pequeno:.4f}")
print(f"p-valor com amostra maior (n=10): {p_grande:.10f}")

# Observe como o p-valor diminui drasticamente quando temos mais dados!
```
