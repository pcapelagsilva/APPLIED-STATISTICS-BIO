**EXERCÍCIO 1:** Teste de Expressão Gênica (T-test)

Cenário: Você está analisando a expressão de um gene específico em duas 
condições celulares: um grupo controle (placebo) e um grupo tratado com uma 
nova droga.

Dados:
- Grupo Controle: [1.2, 1.4, 1.3, 1.5, 1.2, 1.4, 1.3]
- Grupo Tratado: [1.9, 2.1, 2.0, 1.8, 2.2, 1.9, 2.1]

Tarefa:
1. Calcule o p-valor usando um Teste T para amostras independentes.
2. Com um nível de significância de alpha = 0.05, determine se a droga altera 
   significativamente a expressão do gene.

```python
import numpy as np
import pandas as pd
from scipy import stats

controle = [1.2, 1.4, 1.3, 1.5, 1.2, 1.4, 1.3]
tratado = [1.9, 2.1, 2.0, 1.8, 2.2, 1.9, 2.1]

teste_t, p_value = stats.ttest_ind(controle, tratado)

print(f"Estatísticav T: {teste_t:.4f}")

print(f"p-value: {p_value:.4e}")
if p_value < 0.05:
    print("Conclusão: Diferença significativa! A droga altera a expressão!")
else:
    print("Conclusão: Sem diferença significativa.")

    print("\n" + "="*50 + "\n")
```

**EXERCÍCIO 2:** Proporção de Fenótipos (Chi-Square Test)

Cenário: Em um experimento de cruzamento genético, a expectativa teórica 
Mendeliana é de 3:1 (75% fenótipo A, 25% fenótipo B). Você analisou 100 sementes.

Dados Observados:
- Fenótipo A: 68
- Fenótipo B: 32

Tarefa:
1. Realize o teste de Qui-Quadrado para verificar a aderência aos dados esperados.
2. O p-valor indica que o desvio observado é fruto do acaso ou os dados refutam 
   a proporção 3:1?

```python
import numpy as np
import pandas as pd
from scipy import stats

observado = [68, 32]
esperado = [75, 25]

chi_stats, p_value = stats.chisquare(observado, esperado)

print(f"Qui-Quadrado: {chi_stats:.4f}")
print(f"p-value: {p_value:.4f}")
if p_value < 0.05:
    print("Conclusão: Os dados desviam significativamente da proporção 3:1 desejada")
else:
    print("Conclusão: Os dados seguem a proporção esperada (H0 mantida) ")

print("\n" + "="*50 + "\n")
```

**EXERCÍCIO 3:** Correlação de Concentração Proteica (Pearson Correlation)

Cenário: Você deseja verificar se existe uma relação estatística entre o tempo 
de incubação (horas) e a massa de proteína extraída (microgramas).

Dados:
- Tempo (h): [2, 4, 6, 8, 10, 12]
- Proteína (ug): [10, 22, 35, 48, 55, 72]

Tarefa:
1. Calcule o coeficiente de correlação de Pearson e seu respectivo p-valor.
2. Existe uma correlação estatisticamente significante neste experimento?

```python
import numpy as np
import pandas as pd
from scipy import stats

tempo = [2, 4, 6, 8, 10, 12]
proteina = [10, 22, 35, 38, 55, 72]

r, p_value = stats.pearsonr(tempo, proteina)

print(f"Correlação de Pearson (Coeficiente R): {r:.4f}")
print(f"p-value: {p_value:.4e}")
if p_value < 0.05:
    print("Conclusão: Existe uma correlação forte e significativa entre tempo e extração da proteína")
else:
    print("Conclusão: A correlação observada pode ter ocorrida ao acaso")
```

**EXERCÍCIO 4:** Comparação de Múltiplos Grupos (ANOVA)

Cenário: Você está testando três meios de cultura diferentes (A, B e C) para 
verificar qual promove o maior crescimento bacteriano (medido por OD600).

Dados:
- Meio A: [0.82, 0.85, 0.79, 0.81, 0.83]
- Meio B: [0.95, 0.92, 0.98, 0.96, 0.94]
- Meio C: [0.88, 0.86, 0.90, 0.87, 0.89]

Tarefa:
1. Utilize o teste ANOVA de uma via (One-Way ANOVA) para comparar os três grupos.
2. O p-valor indica que pelo menos um dos meios de cultura é significativamente 
   diferente dos outros? (Use alpha = 0.05).

```python
import numpy as np
import pandas as pd
from scipy import stats

grupoa = [0.82, 0.85, 0.79, 0.81, 0.83]
grupob = [0.95, 0.92, 0.98, 0.96, 0.94]
grupoc = [0.88, 0.86, 0.90, 0.87, 0.89]

anova, p_value = stats.f_oneway(grupoa, grupob, grupoc)

print(f"ANOVA: {anova:.4f}")
print(f"p-value: {p_value:.4e}")
if p_value < 0.05:
    print("Conclusão: Pelo menos um dos Grupos de Cultura é significativamente diferente dos demais")
else:
    print("Conclusão: Nenhum dos Grupos de Cultura é significativamente diferente dos outros")

print("\n" + "="*50 + "\n")
```

**EXERCÍCIO 5:** Teste Não-Paramétrico (Mann-Whitney U)

Cenário: Você mediu o tempo de sobrevivência (em dias) de camundongos após um 
tratamento genético. Os dados não parecem seguir uma distribuição normal (são 
assimétricos).

Dados:
- Grupo Controle: [10, 12, 11, 13, 105]  # Nota: O '105' é um outlier
- Grupo Tratado: [15, 18, 16, 20, 22]

Tarefa:
1. Por que usar o Teste T aqui seria arriscado?
2. Execute o teste de Mann-Whitney U para comparar as medianas.
3. Interprete o p-valor obtido.

```python
import numpy as np
import pandas as pd
from scipy import stats

controle = [10, 12, 11, 13, 105]
tratado = [15, 18, 16, 20, 22]

teste_t, p_value = stats.ttest_ind(controle, tratado)
teste_u, p_value = stats.mannwhitneyu(controle, tratado)

print(f"Valor do teste T: {teste_t:.4f}")
print(f"p-value T: {p_value:.4f}")
if p_value < 0.05:
    print("Conclusão: Significativo!! O tratamento genético alterou o tempo de sobrevivência.")
else:
    print("Resultado: Não significativo. Não há evidência de que o tratamento funcione.")
print(f"O valor do teste T ({teste_t})é arriscado porque o outlier 105 'puxa' a média do controle para cima, fazendo com que ela se pareça mais com o grupo tratado do que realmente é; Tornando o p-value maior (menos significativo).")

print("\n" + "-="*50 + "\n")

print(f"Estatística U: {teste_u:.4f}")
print(f"p-value U: {p_value:.4f}")
if p_value < 0.05:
    print("Conclusão: Significativo!! O tratamento genético alterou o tempo de sobrevivência.")
else:
    print("Resultado: Não significativo. Não há evidência de que o tratamento funcione.")
```

**EXERCÍCIO 6:** Teste de Normalidade (Shapiro-Wilk)

Cenário: Antes de escolher entre um Teste T ou um Mann-Whitney, você precisa 
saber se seus dados de expressão proteica seguem uma distribuição normal.

Dados:
- Amostra: [1.2, 1.5, 1.1, 1.4, 1.3, 1.2, 3.5, 0.9, 1.1, 1.4]

Tarefa:
1. Aplique o teste de Shapiro-Wilk.
2. Se o p-valor for < 0.05, os dados NÃO são normais. Qual o resultado aqui?

```python
import numpy as np
import pandas as pd
from scipy import stats

amostra = [1.2, 1.5, 1.1, 1.4, 1.3, 1.2, 3.5, 0.9, 1.1, 1.4]

shapiro_wilk, p_value = stats.shapiro(amostra)

print(f"Shapiro-Wilk: {shapiro_wilk:.4f}")
print(f"p-value: {p_value:.4f}")
if p_value < 0.05:
    print("Conclusão: Os dados NÃO SÃO NORMAIS!!")
else:
    print("Conclusão: Os dados SÃO NORMAIS")

print("\n" + "="*50 + "\n")

import matplotlib.pyplot as plt
import seaborn as sns

# CRIANDO O GRÁFICO COM DOIS SUBPLOTS
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 1. HISTOGRAMA + KDE
sns.histplot(amostra, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribuição da Amostra (Histograma)')
axes[0].set_xlabel('Expressão Proteica')

# 2. BOXPLOT
sns.boxplot(x=amostra, ax=axes[1], color='lightcoral')
axes[1].set_title('Identificação de Outliers (Boxplot)')
axes[1].set_xlabel('Expressão Proteica')

plt.tight_layout()
plt.show()
print("\n" + "="*50 + "\n")
```

**EXERCÍCIO 7:** Múltiplas Comparações (Correção de Bonferroni)

Cenário: Você testou a expressão de 4 genes diferentes simultaneamente. 
Ao rodar os testes T, você obteve os seguintes p-valores:

Dados (p-valores):
- Gene_A: 0.012
- Gene_B: 0.045
- Gene_C: 0.003
- Gene_D: 0.060

Tarefa:
1. Com um alpha inicial de 0.05, quantos genes parecem significativos?
2. Aplique a correção de Bonferroni (alpha_corrigido = alpha / n_testes).
3. Após a correção, quais genes ainda são considerados significativos?

```python
import numpy as np
import pandas as pd
from scipy import stats

p_value = {
    "Gene_A": 0.012,
    "Gene_B": 0.045,
    "Gene_C": 0.003,
    "Gene_D": 0.060
}

alpha_original = 0.05
n_teste = len(p_value)
alpha_corrigido = alpha_original / n_teste

print(f"Alpha Original: {alpha_original}")
print(f"Alpha Corrigido (BonFerroni): {alpha_corrigido:.4f}\n")

print("Avaliação: ")
for gene, p in p_value.items():
    status_original = "SIG" if p < alpha_original else "NÃO SIG"
    status_corrigido = "SIG" if p < alpha_corrigido else "NÃO SIG"
    print(f"{gene}: p={p:.3f} | Antes: {status_original} | Após BonFerroni: {status_corrigido}")
```
