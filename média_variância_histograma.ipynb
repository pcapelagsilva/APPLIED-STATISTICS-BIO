1. MÉDIA + VARIÂNCIA

```python
import numpy as np

dados = np.array([12, 15, 14, 10, 18, 20, 16]) # 

média = np.mean(dados)
variância = np.var(dados)

print(f"A média dos dados é: {média}")
print(f"A variância dos dados é: {variância}")
```

2. HISTOGRAMA

```python
import matplotlib.pyplot as plt
import numpy as np

dados = np.array([12, 14, 10, 18, 20, 16])

plt.hist(dados)
plt.show()
```
